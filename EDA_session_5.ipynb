# finding the outliers

- we already know that outliers available less than 

- step:1 
         -calculate  q1=25p
                    q2=50p
                    q3=75p
- step-2:

        -calculate   IQR =q3-q1

- step-3:

       -calculate lb = q1-1.5*iqr
       -calculate up = q2+1.5*iqr
- step-4:

       -con1=wage data<lb
       -con2=wage data>ub
       -con3=con1 or con2


```python
wage_data=visa_df['prevailing_wage']
q1=np.percentile(wage_data,25)
q2=np.percentile(wage_data,50)
q3=np.percentile(wage_data,75)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
con1=wage_data<lb
con2=wage_data>ub
con3=con1|con2
outliers_data=wage_data[con3]
outliers_data
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming visa_df is your DataFrame and 'prevailing_wage' is the column of interest
wage_data = visa_df['prevailing_wage']

# Calculate the median wage
median_wage = wage_data.median()

# Calculate IQR (Interquartile Range) to detect outliers
Q1 = wage_data.quantile(0.25)
Q3 = wage_data.quantile(0.75)
IQR = Q3 - Q1

# Define outliers as values below (Q1 - 1.5 * IQR) or above (Q3 + 1.5 * IQR)
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with the median value
wage_data_filled = wage_data.apply(lambda x: median_wage if x < lower_bound or x > upper_bound else x)

# Plot the data
plt.figure(figsize=(16, 9))
plt.suptitle('Comparison of Wage Data Before and After Replacing Outliers')

# Original Data
plt.subplot(2, 2, 1)
plt.hist(wage_data, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title('Histogram (Original)')
plt.xlabel('Wage')
plt.ylabel('Frequency')

plt.subplot(2, 2, 2)
plt.boxplot(wage_data, vert=False)
plt.title('Boxplot (Original)')
plt.xlabel('Wage')

# Data with Outliers Replaced
plt.subplot(2, 2, 3)
plt.hist(wage_data_filled, bins=20, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram (Outliers Replaced)')
plt.xlabel('Wage')
plt.ylabel('Frequency')

plt.subplot(2, 2, 4)
plt.boxplot(wage_data_filled, vert=False)
plt.title('Boxplot (Outliers Replaced)')
plt.xlabel('Wage')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout to fit title
plt.show()
```

**How to deal the outliers** 

- 1.we can drop the outliers

   - droping outliers is an easy task, but not recomended
 
   - If you the outliers we are droping the rows
 
   - Which means we are droping the other column values also
 
   - May be some important information we might loose from other column values
 
   - But in  industry if outlier percentage less than 2% of the total
 
   - The number of outliers in pwage is 427
 
   - The number of total observations are 25480
 
   - The percentage is 1.67 which is less than 2%
 
- 2.Fill with Median value

    - We already know that outliers does not affect by median
 
    - So it is the best practice to fill the outliers with median
 
- 3.Capping or florring

    - Instead of fill with median
 
    - We can fill with certain values
 
    - the outliers which are less than  LB fill with LB value
 
    - the outliers which are greater than  UB fill with UB value
 
- **Isolation Forest**

  - Is ML model used to find the outliers and anamolies
